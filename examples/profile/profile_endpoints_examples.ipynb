# 3 -  Write a Profile in a Source from a raw data (resume file) 

```python
# Read a file in a binary form from your disk or download it from the web 

import os 

with open("/home/parsing/Documents/test_resume.pdf", "rb") as f: # <-- Replace with your file path
    binary_profile = f.read()

# Add tags to the profile to be parsed 
tags = [
    {"name": "origin", "value": "linkedin"},
    {"name": "availability", "value": "immediate"},
]
response = client.profile.parsing.add_file(source_key=source_key,
                                            profile_file=binary_profile,
                                            tags=tags,
                                            sync_parsing=1)  # <-- This is optional, to enable sync_parsing reach out to enable it on the source

response
```

# 5 - Edit a Profile already in a Source 

```python
# Edit a profile : edit is a Put not a Patch so you need to provide all the fields

# let's get the profile by reference
reference = "my_custom_reference" # <-- Replace with your profile reference

response = client.profile.storing.get(source_key=source_key, reference=reference)

profile = response['data']

# edit a field or more 
profile['info']['location']['text'] = "Puteaux, France"

# edit the profile
response = client.profile.storing.edit(source_key=source_key,
                                    #    key=profile['key'], # This is Optional, it's kept for backward compatibility with previous versions of the SDK
                                        profile_json=profile)
```
